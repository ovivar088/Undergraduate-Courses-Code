(1) Logistic Regression

```python
#(a)
n = 4
m = 1000

X = np.random.normal(0,1,(m,n))
print(X)
```

```python
#(b)
theta_b = np.array([[1,-1,2,-5]]).T
print(theta_b)
#assuming i is the row indicator of X, y will also be a column vector of size m. Where values will either be 0 or 1 depending on sign(z)
# z = theta * x(i) + 0.5
y_nosign = np.matmul(X,theta_b) + 0.5 #this is questionable
#print(y)

print("\n sign function: \n")
y= np.sign(y_nosign) #this is what we want to generate, the correct labels
print(y) #we will compare this to our generated model using LR
print(np.size(y))
print(y.shape)
```

```python
#Creating Necessary functions for Logisitic Regression

#define sigmoid function
def sigmoid(zeta):
    output = 1 / (1 + np.exp(-zeta))
    return output

#Gradient Descent Functions
def update(X, y, alpha, iteration, w_b):
    m = X.shape[0]
    theta = np.reshape(w_b["theta"],(4,1)) #feautre of w_b --> weights
    bias = w_b["bias"] #feature of w_b --> bias
    for i in range(iteration): 
        sigma_out = sigmoid(np.dot(X, theta) + bias) # sigmoid(theta*x+bias)
        cost = -1/m * np.sum(y * np.log(sigma_out)) + (1 - y) * np.log(1-sigma_out) # 1/m scalar to account for each individual sample

        #GD for Logistic Regression
        dtheta = 1/m * np.dot(X.T, (sigma_out - y))
        theta -= alpha * dtheta #Updating the weights incrementally

        # Accounting for Bias
        db = 1/m * np.sum(sigma_out - y) #raw error to make bias
        bias -= alpha * db #updating biases incrementally

        w_b["theta"] = theta
        w_b["bias"] = bias
        
      #should be printing the final weights
    return w_b #value we want to assign when using this function is weights and biases

```

```python
#(c)
#Executing Gradient Descent/Initialization
alpha = 0.1 / m #step size
iterations = 1000 

init_parameters = {} 
init_parameters["theta"] = np.zeros(X.shape[1])
init_parameters["bias"] = 0

# Define the train function
def train(X,y,alpha,iterations):
    parameters_out = update(X, y, alpha, iterations, init_parameters)
    return parameters_out

 #assigning the new w_b (weights and biases)
# Train the model
parameters_out = train(X, y, alpha, iterations)
print("Weights/Theta: \n", parameters_out['theta'],"\nbias: ", parameters_out['bias'])

# Predict using the trained model
output_values = np.dot(X[:1000], parameters_out["theta"]) + parameters_out["bias"]
predictions = sigmoid(output_values) >= 1/2

for i in range(m):
    if predictions[i]:
        predictions == 1
    else:
        predictions == -1
print(predictions)
```

```python
#(a)
A = np.random.normal(0,1,(3,2))
print(A)

c = np.random.normal(0,3,size =3)
print(c)
```

```python
#(b)

# {x ^ i} ^250 , where each individual x^i 
# x(i) = A*b_i + c , b_i is R^2 --> b_i is two elements
X_3 = np.zeros((250,3))
for i in range(250):
    b = np.random.normal(0,1,(2,)) #because of the for loop a new b is being generated everytime
    X_3[i] = A.dot(b) + c

print("Input data X for problem 3 : \n", X_3)
X_3 = X_3.T
```
